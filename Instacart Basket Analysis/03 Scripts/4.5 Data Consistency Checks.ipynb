The output shows that the column 'days_since_prior_order' has 206,209 missing values. This is likely because these missing values correspond to customers who are placing their very first order. If a user has an order number of 1, there cannot be a "prior order," so the days since the prior order would naturally be missing (NaN).

```python
# Create a new dataframe to clean
df_ords_clean = df_ords

# Create a new column 'new_customer'
# If days_since_prior_order is NaN, it returns True (New Customer)
# If it is not NaN, it returns False (Returning Customer)
df_ords_clean['new_customer'] = df_ords['days_since_prior_order'].isnull() == True

# Check the dataframe to see the new column
df_ords_clean.head()
```

I have chosen to create a new variable that acts as a flag based on the missing value. The missing values in 'days_since_prior_order' indicate that the customer is brand new (it is their first order). I created a new column called 'new_customer' which is set to True if the value is missing, and False otherwise. This allows us to easily identify new customers in our analysis.
