```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

# Create path variable
path = r'C:\Users\howel\OneDrive\Instacart Basket Analysis'

# Import Customer Data
df_customers = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'customers.csv'))

# Check the output
df_customers.head()
```

```python
# Rename columns to be consistent with other dataframes
df_customers.rename(columns = {'First Name': 'first_name', 'Surnam': 'last_name', 'Gender': 'gender', 'STATE': 'state', 'Age': 'age', 'date_joined': 'date_joined', 'n_dependants': 'n_dependants', 'fam_status': 'fam_status', 'income': 'income'}, inplace = True)

# Check the changes
df_customers.head()
```

```python
# Check for mixed types
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].map(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len(df_customers[weird]) > 0:
    print (col)
```

```python
# Check for missing values
df_customers.isnull().sum()
```

```python
# Check for duplicates
df_customers_dups = df_customers[df_customers.duplicated()]
df_customers_dups
```

```python
# 1. Import most recent main data
orders_products_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged.pkl'))

# 2. Check data types of the key column 'user_id'
orders_products_merged['user_id'].dtype
df_customers['user_id'].dtype

# 3. Merge the dataframes
orders_products_cust_merged = orders_products_merged.merge(df_customers, on = 'user_id', how = 'inner')

# 4. Check the shape to ensure no data was lost significantly
orders_products_cust_merged.shape
```
